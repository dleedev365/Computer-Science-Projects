# Extract text from 'features'

# 1. Word Count

```python
#tokenize words
#each phrase is considered as a word
indexes = list(df.index.values.tolist())
values = list(df['features'][indexes])
text_list = []
holder = 0
for i in range(len(values)):
    for ele in values[i]:
        if(len(ele)>50):
            temp = []
            temp = re.split(r'[**]', ele)
            for j in range(len(temp)):
                text_list.append(temp[j])  
        else:
            text_list.append(ele)
        holder += 1   
Counter(text_list)
```

```python
#create 2-d list to store words 
#without duplicate and the number of a word 
text_list1 = list(dict.fromkeys(text_list))
dic_list = [[0 for x in range(len(text_list1))] for y in range(2)] 


for i in range(len(text_list1)):
    temp = text_list1[i]
    dic_list[0][i] = temp
    dic_list[1][i] = (text_list.count(temp))

#print(dic_list[0][0])
#print(int(dic_list[1][0]))
print(dic_list[0:10])
```

# 2. Word Freq

```python
#create 2-d list to store words 
#without duplicate and the frequency of a word 
total_word = 0
for i in range(len(text_list1)):
    total_word += dic_list[1][i]

freq_list = [[0 for x in range(len(text_list1))] for y in range(2)] 
for i in range(len(text_list1)):
    temp = text_list1[i]
    freq = dic_list[1][i]/total_word
    freq_list[0][i] = temp
    freq_list[1][i] = freq
    
#print(freq_list[0][1])
#print(freq_list[1][1])
print(freq_list[0:10])
```
